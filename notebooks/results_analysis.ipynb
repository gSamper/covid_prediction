```python
import os
import plotly.express as px
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
import pandas as pd

# Load the data as a pandas DataFrame
file_path = "data/training_data/training_datav2_0-weekly.pkl"
training_data = pd.read_pickle(file_path)

best_countries = ["Finland", "South Korea", "Germany", "Latvia", "Spain"]
worst_countries = ["Greece", "Malta", "Bolivia", "Ireland", "Iceland"]


df_best = training_data[training_data["location"].isin(best_countries)]
df_worst = training_data[training_data["location"].isin(worst_countries)]
```

```python

# Filter the data for the best countries
df_best_filtered = training_data[training_data["location"].isin(best_countries)]

scaler = MinMaxScaler()

# Scale the data for each country individually
df_best_filtered["scaled_weekly_cases"] = df_best_filtered.groupby("location")["weekly_cases"].transform(
    lambda x: scaler.fit_transform(x.values.reshape(-1, 1)).flatten()
)

# Plot 1: Weekly Cases
fig_weekly_cases = px.line(
    df_best_filtered,
    x="date",  # Replace with the actual column name for dates in your DataFrame
    y="scaled_weekly_cases",
    color="location",  # Different color for each country
    facet_col="location",  # Create a small plot for each country
    facet_col_wrap=3,  # Wrap subplots into rows of 3
    title="Weekly Cases for Best Countries",
    labels={"scaled_weekly_cases": "Weekly Cases", "date": "Date", "location": "Country"}
)

# Show the first plot
fig_weekly_cases.show()

# Plot 2: Additional Metrics (C, E, G, S)
columns_to_plot = ["C", "E", "G", "S"]

# Reshape the DataFrame to a long format for additional metrics
df_long_metrics = pd.melt(
    df_best_filtered,
    id_vars=["date", "location"],  # Columns to keep
    value_vars=columns_to_plot,   # Columns to unpivot
    var_name="metric",            # Name for the new column indicating the variable
    value_name="value"            # Name for the new column with the values
)

fig_additional_metrics = px.line(
    df_long_metrics,
    x="date",  # Replace with the actual column name for dates in your DataFrame
    y="value",
    color="metric",  # Different color for each metric (C, E, G, S)
    facet_col="location",  # Create a small plot for each country
    facet_col_wrap=3,  # Wrap subplots into rows of 3
    title="Additional Metrics (C, E, G, S) for Best Countries",
    labels={"value": "Value", "date": "Date", "location": "Country", "metric": "Metric"}
)

# Show the second plot
fig_additional_metrics.show()
```

```python
df_worst_filtered = training_data[training_data["location"].isin(worst_countries)]

scaler = MinMaxScaler()

# Scale the data for each country individually
df_worst_filtered["scaled_weekly_cases"] = df_worst_filtered.groupby("location")["weekly_cases"].transform(
    lambda x: scaler.fit_transform(x.values.reshape(-1, 1)).flatten()
)
# Create a faceted line plot
fig = px.line(
    df_worst_filtered,
    x="date",  # Replace with the actual column name for dates in your DataFrame
    y="scaled_weekly_cases",
    color="location",  # Different color for each country
    facet_col="location",  # Create a small plot for each country
    facet_col_wrap=3,  # Wrap subplots into rows of 3
    title="Weekly Cases for Best Countries (Faceted)",
    labels={"scaled_weekly_cases": "Weekly Cases", "date": "Date", "location": "Country"}
)

# Show the plot
fig.show()
```
