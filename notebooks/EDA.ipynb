````python
def get_date_range(df:pd.DataFrame, column_name = 'date') -> None:
    """
    Get the date range of a DataFrame column containing dates.

    Args:
        df (pd.DataFrame): A pandas dataframe containing a date column.
        column_name (str, optional): Name of the column which contains the dates. Defaults to 'date'.
    """
    if type(df[column_name].iloc[0]) != pd.Timestamp:
        df[column_name] = pd.to_datetime(df[column_name])
    min_date = df[column_name].min()
    max_date = df[column_name].max()
    print(f"Date range: {min_date.date()} to {max_date.date()}")
````

#### Load  and adapt data

````python
###############################################################
hospital_data = pd.read_pickle('data/pickles/hospital_data.pkl')
cases_data = pd.read_pickle('data/pickles/cases_data.pkl')

hospital_data = hospital_data.pivot(index = ['entity', 'iso_code', 'date'],columns='indicator', values='value')
hospital_data = hospital_data.reset_index()
````

````python
folder_path = r'data\timeseries_pickles'

# Explicit mapping of file names to variable names
file_to_variable_map = {
    "ContainmentHealthIndex.pkl": "c_index",
    "EconomicSupportIndex.pkl": "e_index",
    "GovernmentResponseIndex.pkl": "g_index",
    "StringencyIndex.pkl": "s_index"
}

# Dictionary to store the loaded DataFrames
pickles_dict = {}

# Load the pickles and assign them to variables dynamically
for file_name, variable_name in file_to_variable_map.items():
    file_path = os.path.join(folder_path, file_name)
    if os.path.exists(file_path):  # Check if the file exists
        pickles_dict[file_name] = pd.read_pickle(file_path)
        # globals()[variable_name] = pickles_dict[file_name]
    else:
        print(f"Warning: {file_name} not found in {folder_path}")
````

````python
c_index = pickles_dict['ContainmentHealthIndex.pkl']
e_index = pickles_dict['EconomicSupportIndex.pkl']
g_index = pickles_dict['GovernmentResponseIndex.pkl']
s_index = pickles_dict['StringencyIndex.pkl']
policies_datasets = [c_index, e_index, g_index, s_index]
````

#### Check and clean available countries

````python
# Get the common countries in all the policy datasets and the cases dataset
policy_countries = set(c_index.CountryName).intersection(e_index.CountryName, g_index.CountryName, s_index.CountryName)
cases_countries = set(hospital_data.entity).intersection(cases_data.location)
all_countries = policy_countries.intersection(cases_countries)

#explanation
print(f"Countries in policy data: {len(policy_countries)}")
print(f"Countries in cases data: {len(cases_countries)}")
print(f"Total Available countries: {len(all_countries)}")
print(list(all_countries))
````

````python
#take only the countries that are in all datasets
cases_clean = cases_data.loc[cases_data.location.isin(all_countries)].copy()
hospital_clean = hospital_data.loc[hospital_data.entity.isin(all_countries)].copy()

#convert the date column to datetime format
cases_clean['date'] = pd.to_datetime(cases_clean['date'])
hospital_clean['date'] = pd.to_datetime(hospital_clean['date'])

#merge the two dfs by time and location
target_data = pd.merge(cases_clean, hospital_clean, how='inner', 
                       left_on=['location', 'date'], right_on=['entity', 'date']).drop(columns='entity')
# Drop duplicate columns after merging
# target_data = target_data.loc[:, ~target_data.columns.duplicated()]

target_data.head()
````
