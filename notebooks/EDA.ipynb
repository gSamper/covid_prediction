```python
#Data from OWID
hospital_data = pd.read_csv('data\covid-hospitalizations.csv')
cases_data = pd.read_csv('data\cases_full.csv')
vaccine_data = pd.read_csv(r'data\vaccinations.csv')
# vaccine_by_age_data = pd.read_csv(r'data\vaccinations-by-age-group.csv')

# Data from OxCGRT
policy_data = pd.read_csv('data\OxCGRT_compact_national_v1.csv')
policy_vaccine_data = pd.read_csv(r'data\OxCGRT_vaccines_full_national_v1.csv')
```

```python
hospital_data
sweden = hospital_data[hospital_data['entity'] == 'Sweden']

# sweden = sweden.set_index('date')
temp = sweden.pivot(index = ['entity', 'iso_code', 'date'],columns='indicator', values='value')
temp = temp.reset_index()
temp.head()
```

```python

# Melt the DataFrame to have a long format for easier plotting
temp_melted = temp.melt(id_vars=['date'], 
                        value_vars=['Daily ICU occupancy', 
                                    'Daily ICU occupancy per million', 
                                    'Daily hospital occupancy', 
                                    'Daily hospital occupancy per million'],
                        var_name='Indicator', 
                        value_name='Value')

# Create the Plotly line chart
fig = px.line(temp_melted, 
              x='date', 
              y='Value', 
              color='Indicator', 
              title='Daily ICU and Hospital Occupancy Trends',
              labels={'date': 'Date', 'Value': 'Occupancy'},
              template='plotly')

# Update layout for better readability
fig.update_layout(xaxis_title='Date', 
                  yaxis_title='Occupancy', 
                  legend_title='Indicator',
                  xaxis=dict(showgrid=False),
                  yaxis=dict(showgrid=True))

# Show the plot
fig.show()
```

```python
candidate_countries = set()

candidate_countries.update(cases_data['location'].unique())
candidate_countries.intersection_update(vaccine_data['location'].unique())
# candidate_countries.intersection_update(vaccine_by_age_data['location'].unique())
candidate_countries.intersection_update(hospital_data['entity'].unique())

candidate_countries.intersection_update(policy_data['CountryName'].unique())
candidate_countries.intersection_update(policy_vaccine_data['CountryName'].unique())
print(len(candidate_countries))
print(candidate_countries)
```
